```python
import numpy as np
import matplotlib.pyplot as plt
from functools import partial

def zero_noise(dx):
    return 0

def uniform(a, b, dx):
    return np.random.uniform(a, b, size=dx.shape)

def percent_noise(dx, alpha=.5):
    return np.random.uniform(-alpha/2*dx, +alpha/2*dx)
    

def simulate(deltas, noise_fn=zero_noise):
    
    path = np.zeros(deltas.size+1)
    path[0] = 0
    for i in range(1, len(path)):
#         print(noise_fn(deltas[i-1]))
        path[i] = path[i-1] + deltas[i-1] + noise_fn(deltas[i-1])
    
    return path
```

```python
fig = plt.figure(figsize=(10,7))

# deltas = np.random.uniform(-1,1,size=100)
# deltas = np.linspace(0, 10, 100)
deltas = np.ones(50)
path_true = simulate(deltas);


simulated_paths = []
N = 20
for i in range(N):
    path_simulated = simulate(deltas, percent_noise)
    
    simulated_paths.append(path_simulated);
    plt.plot(path_simulated, 'm', alpha=0.2)

true_paths = np.array(true_paths)
simulated_paths = np.array(simulated_paths)

plt.plot(path_true, 'bo-')
```

```python
fig, ax1 = plt.subplots()

# Plot the absolute value of the error over the course of the path
error = np.mean(np.abs(simulated_paths - path_true), axis=0)[1:]
ax1.plot(error, 'r.-')

ax2 = ax1.twinx()
ax2.plot(deltas, '.')
```
